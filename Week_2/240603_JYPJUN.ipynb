## 1. AAS 이란?

##### - AAS(Asset Administration Shell)은 컴포넌트, 기계 또는 생산 라인과 같은 자산의 표준화된 디지털 표현으로, 다른 디지털 시스템과의 원활한 통신 및 상호작용을 하기위한 일종의 프로파일을 의미함

   
     
   
- AAS는 산업 4.0(Industrie 4.0)에서 중요한 개념으로, 물리적 자산(Asset)과 디지털 세계를 연결하는 디지털 트윈(Digital Twin) 구현의 핵심 요소이다. AAS는 각 자산에 대한 표준화된 디지털 표현을 제공하며, 이를 통해 자산의 데이터 및 기능을 디지털화하고 관리할 수 있다.

##### 1. AAS의 주요 요소와 개념
   
     
     
   
  - 자산(Asset)
    - 자산은 물리적 장비, 기계, 센서, 소프트웨어 등 물리적 또는 논리적인 개체를 의미

  - 관리 쉘(Administration Shell)
    - 관리 쉘은 자산의 디지털 표현을 의미, 이는 자산의 데이터, 기능, 서비스 등을 포함하는 메타데이터의 구조체
    - 관리 쉘은 자산에 대한 모든 디지털 정보를 담고 있으며, 자산의 디지털 트윈 역할이다.
  
  - 디지털 트윈(Digital Twin)
    - 디지털 트윈은 물리적 자산의 디지털 복제본으로, 자산의 상태, 동작, 데이터를 실시간으로 반영
    - AAS는 디지털 트윈의 구현을 가능하게 하며, 이를 통해 자산의 성능을 모니터링하고 최적화할 수 있다.

##### 2. AAS의 구조
- AAS는 여러 서브 모델(Submodel)로 구성되며, 각 서브 모델은 자산의 특정 측면에 대한 정보를 제공

  1. 아이덴티티(Identity)
    - 자산의 고유 식별 정보를 포함
    - 자산의 일련 번호, 모델 번호, 제조업체 정보 등이 포함
  
  2. 서브 모델(Submodel)   

         
       
    - 자산의 다양한 속성 및 특성에 대한 세부 정보를 담고 있음
    - 각 속성간의 상호 운용성을 설명하기 위해 각 관점/기술 영역에 대한 단일 그룹(Submodel)을 만드는 것이 주 목적
    - 예 : 기계적 특성 서브 모델, 전기적 특성 서브 모델, 유지보수 서브 모델 등
  
  3. 속성(Properties)
    - 각 서브 모델 내의 속성은 자산의 구체적인 데이터를 포함
    - 예 : 온도, 압력, 전류 등의 센서 데이터, 기계 부품의 규격, 유지보수 기록 등

##### 3. AAS의 기능 및 활용

  1. 데이터 통합 및 상호운용성
  2. 실시간 모니터링 및 예측 유지보수
  3. 디지털 트윈 구축
  4. 수명 주기 관리


##### 4. AAS의 장점

  1. 효율성 증가   
    - 자산의 디지털화를 통해 운영 효율성을 증가시킬 수 있음   
    - 실시간 데이터 모니터링과 분석을 통해 운영상의 결정을 신속하게 내릴 수 있음
  
  2. 비용 절감   
    - 예측 유지보수와 최적화를 통해 유지보수 비용과 운영 비용을 절감   
    - 불필요한 다운타임을 줄이고 자산의 수명을 연장

## 2. 데이터 품질 전처리

## 1. 데이터 전처리란?

- 데이터를 분석 및 처리에 적합한 형태로 만드는 과정을 총칭
- 데이터 전처리는 데이터 분석 및 처리과정에서 중요한 단게이며 데이터 분석, 데이터 마이닝, 머신 러닝 프로젝트에 적용
- 일반적으로 데이터는 비어있는 부분이 있거나 정합성이 맞지 않는 경우가 많다.

     
  
  - 데이터 분석 개발자는 시간의 60%를 데이터 정리 및 구성하는 데에 사용
  - 데이터 세트 수집은 19%를 차지함

##### 1. 데이터 전처리 과정
- 데이터 수집 -> 데이터 정제 -> 데이터 변환 -> 데이터 통합 및 분할 -> 데이터 저장 및 문서화

1. 데이터 수집

    - 데이터 소스 식별 및 수집과정을 거친다. 데이터베이스, 웹 스크래핑, API, 파일 등 다양한 소스로부터 데이터를 수집

2. 데이터 정제

    - 결측값 처리(Missing Value Handling): 결측값을 제거하거나 대체합니다. 대체 방법에는 평균, 중앙값, 최빈값 대체 또는 예측 모델을 통한 대체가 있음

    - 이상값 처리(Outlier Handling) : 이상값을 식별하고 제거하거나 수정

    - 중복 데이터 제거(Duplicate Removal) : 중복된 데이터를 식별하고 제거

3. 데이터 변환

    - 데이터 타입 변환(Data Type Conversion) : 적절한 데이터 타입으로 변환 ex) 날짜 -> 문자열(datetime)

    - 스케일링 및 정규화(Scaling and Normalization) : 데이터의 스케일을 조정하여 모델링에 적합하게 변환 (Min-Max 스케일링, Z-score etc..)

    - 데이터 인코딩(Data Encoding) : 범주형 변수를 수치형으로 변환 (One-Hot Encoding, Label Encoding etc..)

    - 특성 생성(Feature Engineering) : 기존 데이터를 바탕으로 새로운 특성을 생성 ex) 날짜 데이터 -> 요일, 월, 계절 추출

    - 특성 선택(Feature Selection) : 분석 또는 모델링에 중요하지 않은 변수 제거

4. 데이터 통합 및 분할

    - 데이터 통합(Data Integration) : 여러 소스에서 수집된 데이터를 통합하여 하나의 데이터셋으로 생성

    - 데이터 분할(Data Splitting) : 데이터를 학습용(train), 검증용(validation), 테스트용(test)으로 분할

5. 데이터 저장 및 문서화

    - 데이터 저장(Data Storage) : 전처리된 데이터를 분석 및 모델링에 사용할 수 있도록 저장 (Csv, 데이터베이스, HDF5 etc..)

    - 문서화(Documenting) : 데이터 전처리 과정을 문서화하여 재현 가능성을 높인다.

## 2. KAMP 데이터셋에서의 데이터 품질 전처리

- Raw Data의 수집은 다양한 센서를 사용하여 배터리의 수명과 관련이 있는 데이터를 수집 및 저장

- 실제 현장의 데이터의 의미 없는 값을 포함한 값이나, 결측치 혹은 이상 데이터는 제거

- 아래 6가지 기준으로 데이터 품질 지수를 산출 및 데이터의 전처리를 통하여 품질 개선
   
  
   
   
1. 완전성 품질지수   

        
  - 결측 데이터의 값이 30% 이상인 레코드는 데이터의 완전성을 떨어뜨리기 떄문에 각 열에 대한 결측 데이터값의 비율을 확인 후 삭제
  - 총 결측 데이터의 개수를 확인하기 위해 'isnull()'함수 사용, 각 열에서의 결측 데이터의 개수를 구한 뒤 'sum()'으로 합산

2. 유일성 품질지수
   
        
  - 데이터에서 유일한 값을 갖는 열을 찾는다. 시계열 데이터는 Timestamp를 사용
  - 유일한 값을 갖는 열에 대해 유일한 데이터 값을 확인 및 개수를 구함
  - ex) 데이터 테이블에서 같은 시간대에 중복해서 기록되었다면 유일성에 위배

3. 유효성 품질지수

       
  - 데이터가 주어진 유효범위를 벗어나는지 확인
  - 전체 데이터의 개수와 유효범위 안에 있는 데이터의 개수를 구함

4. 일관성 품질지수

       
  - 데이터 테이블 간 종속관계를 확인
  - 데이터의 타입이 정해진 형식에 맞는지 확인 및 전체 데이터의 개수와 형식에 맞는 데이터의 개수를 구함

5. 정확성 품질지수

       
  - 데이터 항목에 대하여 검증된 정확한 정보(참조데이터)가 있는지 확인
  - 참조 데이터가 있는 경우, 데이터 항목 값과 참조 데이터의 값을 비교
  - 차이가 0이 아니면 정확성을 위배하는 데이터가 존재

6. 무결성 품질지수

       
  - 무결성 품질지수는 앞서 구한 유일성, 유효성, 일관성 지수를 기초로 위의 식으로 계산

```python
# 품질 전처리에 사용하는 라이브러리
import pandas as pd
import numpy as np
from datetime import date
from datetime import time
from datetime import datetime
from dateutil.parser import parse
# parse는 다양한 형식의 날짜와 시간을 자동으로 인식하여 datetime 객체로 변환해주는 함수
# 해당 함수는 매우 유연하며, 다양한 형식의 문자열 날짜를 파싱할 수 있다.

# 다양한 형식의 날짜 문자열을 파싱
# date1 = parse("2023-06-01") # 2023-06-01 00:00:00
# date2 = parse("June 1, 2023") # 2023-06-01 00:00:00
# date3 = parse("01/06/2023") # 2023-06-01 00:00:00
# date4 = parse("2023/06/01 12:30:45") # 2023-06-01 12:30:45

# parse는 모호한 날짜 형식 (01/02/2023) 와 같은 일,월,년 인지 월,일,년 인지에 대한 처리가 애매할 때 사용
# dayfirst 인자를 사용하여 일을 먼저 해석할지 여부를 지정할 수 있음
# parse("01/02/2023", dayfirst=True) # 2023-02-01 00:00:00 -> True일 경우 두 번째 값이 월


file_path = './1000_chg.csv' # 데이터 출처 www.kamp-ai.kr
data = pd.read_csv(file_path)
data # Data Sheet
```

```python
# 품질 전처리를 위한 문자열이 date, time 형식인지 체크 함수 정의
def is_date(string, fuzzy=False):
    try:
        parse(string, fuzzy=fuzzy)
        return True
    except ValueError:
        return False

def string_is_time(string):
    try:
        datetime.strptime(string, '%H:%M:%S')
        return True
    except ValueError:
        return False
```

```python
# 완전성 품질지수
# 총 결측 데이터의 개수를 확인하기 위해 'isnull()'함수를 사용
# 각 열에서의 결측 데이터 개수를 구한 뒤 'sum()'함수를 이용하여 합산
# getCompleteness()는 완전성 품질지수를 구하는 함수
def getCompleteness(data):
    total_null_values = data.isnull().sum().sum() # null 의 개수
    total_notNull_values = data.count().sum() # 실존하는 값의 개수
    total_values = total_null_values + total_notNull_values # 전체 합
    completeness = (total_notNull_values / total_values) * 100
    return total_null_values, completeness

total_null_values, completeness = getCompleteness(data)
print(f"Total null values = {total_null_values} values")
print(f"완전성 품질지수 : {completeness:.2f}%")
```

```python
# 위 함수의 완전성 품질지수는 100%로 결측데이터가 없음
# 다만, 결측치가 존재한다면 아래의 함수를 사용하여 결측 데이터를 처리 후 다시 확인
# isnull : 데이터프레임에서 결측치(null)을 찾음
# iloc : 위치 기반 인덱싱을 사용하여 데이터프레임의 특정 위치에 있는 값을 접근
# notna : 결측치가 아닌 값을 찾음
# idxmax : 최대값의 인덱스를 반환
# iat : iloc과 비슷하지만 iat는 단일 값에 대하여 접근하며 해당 셀에 대한 값을 바꿈
# drop : 데이터프레임에서 특정 행이나 열을 삭제할 때 사용 axis=0은 행, 1은 열
# fillna : fillna(value, inplace=True) -> 결측값을 value로 바꿈
    # mode() 메서드는 열에서 가장 빈도가 높은 값을 반환
    # 아래에서 쓰인 fillna(df[i].mode()[0], inplace=True)는 df i번째 열의 최빈값을 기준으로 변경하고 있음
    

import math

def handleMissingValue(data):
    df = data.copy()
    numeric_df = df.columns[df.dtypes != 'object'] # ?? 얘는 왜 쓰는거지??
    categorical_df = df.columns[df.dtypes == 'object']
    for i in categorical_df:
        df[i].fillna(df[i].mode()[0], inplace=True)
    
    # null의 값을 가지고 있는 columns만 제거
    for col in df.columns:
        if df[col].isnull().all():
            df = df.drop([col], axis=1) # null값의 열을 삭제하고 그 값을 다시 반환
    
    df_flag_null = df.isnull()
    i, c = np.where(df_flag_null) # df에서 null인 값을 찾아서 행과 열을 i와 c에 기록
    
    # ???
    for j in range(len(i)): # 행의 수 만큼 반복
        if i[j] ==0:
            s = df.iloc[:, c[j]]
            id_s = s.notna().idxmax()
            val = df.iat[id_s, c[j]]
            df.iat[i[j], c[j]] = val
            
        elif i[j] == (len(df)-1):
            print("end")
            print(i[j], c[j])
            s = df.iloc[:, c[j]]
            id_s = s.notna()[::-1].idxmax()
            print(id_s, c[j])
            print(val)
            df.iat[i[j], c[j]] = val
        
        else:
            low = df.iat[i[j]-1, c[j]]
            high = df.iat[i[j]+1, c[j]]
            if (math.isnan(high)):
                val = low
            else:
                val = (low + high) / 2
                df.iat[i[j], c[j]] = val
    return df

print(handleMissingValue(data))
```

```python
# 유일성 품질지수
time_series = data["Time"]
def getUniqueness(data):
    total_data = data.count()
    unique = len(data.unique())
    uniqueness = (unique / total_data) * 100
    return uniqueness

print(f"유일성 품질지수: {getUniqueness(time_series)}%")
```

```python
# 유효성 품질지수
# 데이터의 유효 범위를 range_min과 range_max 값으로 지정, 유효한 범위 내에 속하는 데이터를 count
# loc : pandas 라이브러리 중 슬라이싱 메서드 / 
def getSeriesValidity(data, range_min, range_max):
    column_name = data.name
    total_data = data.count()
    
    if (len(column_name) >= 4 and column_name[0:4] == 'Time'):
        # time_conv = pd.to_datetime(data).dt.time # 아래에 format을 사용하는게 시간이 덜 걸림
        time_conv = pd.to_datetime(data, format='%H:%M:%S').dt.time
        temp = data.loc[time_conv >= range_min] # ???
        temp = data.loc[time_conv <= range_max] # ??? 왜 이런식으로 할당함? 이해 불가..
        # 아래처럼 해야하는거 아님? 뒤에도 다 똑같이 바꿔야할거같은데..
        # temp = data.loc[(time_conv >= range_min) & (time_conv <= range_max)]
        in_range_data = temp.count()
    
    elif (len(column_name) >= 4 and column_name[0:4] == 'Date'):
        date_conv = pd.to_datetime(data).dt.date
        temp = data.loc[date_conv >= range_min]
        temp = data.loc[date_conv <= range_max]
        in_range_data = temp.count()
    
    else: # 위에는 이상하게 했으면서 아래는 왜 또 정상임?
        in_range_data = data.loc[(data >= range_min) & (data <= range_max)].count()
    
    validity = in_range_data / total_data * 100
    return validity

def getValidity(data, ranges):
    validities = []
    for col in ranges:
        current_validity = getSeriesValidity(data[col], ranges[col][0], ranges[col][1])
        validities.append(current_validity)
    average_validity = sum(validities) / len(validities)
    return average_validity

columns = data.columns
range_dictionary = {}
default_range = (-200, 200)
set_range = {
    "Voltage": (0, 1000),
    "Date": (date(1999, 1, 1), date(2023, 12, 12)),
    "Time": (time(0, 0), time(23, 59)),
    "SerialNumber": (0, 1000),
}

for column_name in columns: # columns 다 돌면서 조건 추가
    if column_name in set_range:
        range_dictionary[column_name] = set_range[column_name]
    else:
        range_dictionary[column_name] = default_range
    # print(range_dictionary)

print(f"유효성 품질지수 : {getValidity(data, range_dictionary):.2f}%")
```

```python
# 일관성 품질지수
# 전체 데이터의 개수와 형식에 맞는 데이터의 개수를 구하여 일관성 지수를 계산
def getSeriesConsistency(data, data_type):
    total_data = data.count()
    
    if (data.name == 'Date'):
        types = list(map(lambda item: is_date(item), data))
    elif (data.name == 'Time'):
        types = list(map(lambda item: string_is_time(item), data))
    else:
        types = list(map(lambda item: isinstance(item, data_type), data))
    
    types = pd.Series(types)
    consistent_data = types.sum()
    consistency = (consistent_data / total_data) * 100
    return consistency

def getConsistency(data, types):
    consistent_data = []
    for col in types:
        current_const = getSeriesConsistency(data[col], types[col])
        consistent_data.append(current_const)
    average_const = sum(consistent_data) / len(consistent_data)
    return average_const

columns = data.columns
const_dictionary = {}
default_type = (float, int)
numeric_type = (float, int)
set_types = {
    "Voltage": numeric_type,
    "Current": numeric_type,
    "Power": numeric_type,
}

for column_name in columns:
    if column_name in set_types:
        const_dictionary[column_name] = set_types[column_name]
    else:
        const_dictionary[column_name] = default_type

print(f"일관성 품질지수 : {getConsistency(data, const_dictionary):.2f}%")
```

## 3. PCA 란?

##### 1. PCA 
- [참고] : https://ddongwon.tistory.com/114
- PCA 알고리즘의 원리는 위의 블로그에서 이해하는 것을 추천, 세부 내용은 많이 생략하였음

  - 주성분 분석(Principal Component Analysis)은 고차원 데이터를 저차원으로 변환하는 데 사용되는 통게적 기법

  - 데이터의 변동성을 최대한 보존하면서 데이터를 요약하는 데 중점을 둔다.

    

  - 2차원의 데이터-셋을 1차원으로 줄이는 방법으로 x, y 좌표로 구성된 2차원 데이터를 1차원의 선으로 사영(Projection)한다.

  - 다만, 2차원의 데이터의 특징을 모두 살리면서 1차원의 데이터로 바꿀 수는 없다.

  - 따라서, 최대한 특징을 살리면서 차원을 낮춰주는 PCA 알고리즘 

1. 목적

  - 차원 축소 : 고차원 데이터를 저차원으로 축소하여 처리 속도를 향상시키고, 시각화를 용이하게 한다.

  - 특성 추출 : 데이터의 중요한 특성을 추출하여 데이터의 이해와 해석을 돕는다.

  - 노이즈 제거 : 데이터에서 노이즈를 제거하고, 신호 대 잡음 비율을 개선한다.

  - 데이터 압축 : 데이터의 크기를 줄여 저장 공간을 절약하고, 효율적인 데이터 전송을 가능하게 한다.

2. PCA 원리 방식

    1. 각 축에 대한 평균값을 구한 뒤, 해당 점이 원점이 되도록 Shift 해준다.

        

    2. 데이터에서 원점을 지나는 직선에 수선의 발을 내려, 해당 길이가 최대가 되는 직선을 찾는다.

        

    3. 찾은 직선을 PC1으로 설정하고, loading score를 구한다.

        

    4. PC1에 직교하는 직선을 PC2로 잡는다.

        

    5. PC1과 PC2를 축으로 하여 회전시킨 뒤, scree plot을 생성한다.
